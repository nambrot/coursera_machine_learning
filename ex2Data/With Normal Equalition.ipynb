```python
import numpy as np
m = 50
raw_x = np.loadtxt('ex2x.dat').astype(np.float64).reshape((m, 1))
raw_y = np.loadtxt('ex2y.dat').astype(np.float64).reshape((m, 1))

xx = np.concatenate((np.ones((m, 1)), raw_x), axis = 1)
```

```python
theta = np.dot(np.dot(np.linalg.inv(xx.T.dot(xx)), np.transpose(xx)), raw_y)
theta
```

```python
%matplotlib inline
import matplotlib.pyplot as plot
plot.plot(xx.dot(theta))
plot.plot(raw_y)
```
