Linear Regression

Below is an implementation of a Linear Regression with a batch gradient descent.

```python
import pandas as pd
from __future__ import print_function

import tensorflow as tf

iterations = 1500

ages = pd.read_csv("ex2x.dat", header=None)
ages.columns = ['Age']
heights = pd.read_csv("ex2y.dat", header=None)
heights.columns = ['Height']

with tf.Session() as sess:
    m = tf.constant(50.0, dtype=tf.float64)
    inputs = tf.concat(1, [tf.ones([50, 1], tf.float64), tf.constant(ages.values)])
    outputs = tf.constant(heights.values)
    theta = tf.Variable(tf.zeros([1, 2], tf.float64))
    tf.initialize_all_variables().run()

    alpha = tf.constant(0.07, dtype=tf.float64)
    multiplier = tf.truediv(alpha, m)
    
    for i in range(iterations):
        cost = tf.sub(tf.matmul(inputs, tf.transpose(theta)), outputs)
        sess.run(tf.assign(theta, tf.sub(theta, tf.transpose(tf.mul(multiplier, tf.matmul(tf.transpose(inputs), cost))))))

    print("Final Theta: ", theta.eval())
    
    predictions = pd.DataFrame(tf.matmul(inputs, tf.transpose(theta)).eval())
    predictions.columns = ['Prediction']
    
    %matplotlib inline
    pd.concat([ages, heights, predictions], axis=1).set_index('Age').plot()
```

Once the model was trained and the parameters are returned, we can use it to predict new values

```python
def predict(age):
    return 0.75015039 + age * 0.06388338

print("Predicted height for age", 3.5, "is", predict(3.5))
print("Predicted height for age", 7, "is", predict(7))
    
```

```python
import pandas as pd
import numpy as np
from __future__ import print_function
import matplotlib.pyplot as plt
import tensorflow as tf

ages = pd.read_csv("ex2x.dat", header=None, dtype=np.float32)
ages.columns = ['Age']
heights = pd.read_csv("ex2y.dat", header=None, dtype=np.float32)
heights.columns = ['Height']
```

```python
W = tf.Variable(tf.zeros([1]))
b = tf.Variable(tf.zeros([1]))
y = W * np.array(ages.values).astype(np.float32) + b

loss = tf.reduce_mean(tf.square(y - np.array(heights.values).astype(np.float32)))
optimizer = tf.train.GradientDescentOptimizer(0.009)
train = optimizer.minimize(loss, var_list=[W, b])

init = tf.initialize_all_variables()

# Launch the graph.
sess = tf.Session()
sess.run(init)

for step in range(3801):
    sess.run(train)
    if step % 500 == 0:
        pred = pd.DataFrame(sess.run(y))
        pred.columns = ['Prediction']
        %matplotlib inline
        pd.concat([ages, heights, pred], axis=1).set_index('Age').plot()
        plt.show()
```
