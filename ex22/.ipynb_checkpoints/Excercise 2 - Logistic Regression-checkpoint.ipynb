# 1. College admission

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

m = 100
n = 2
alpha = 0.0011

data = np.loadtxt(open("ex2data1.txt","rb"),delimiter=",")
admitted = data[data[:, 2] == 1, :]
not_admitted = data[data[:, 2] == 0, :]
plt.plot(admitted[:, 0], admitted[:, 1], 'x')
plt.plot(not_admitted[:, 0], not_admitted[:, 1], 'o')
```


### Plain Logistic Regression

```python
x = np.concatenate((np.ones((m, 1)), data[:, 0:2]), axis = 1)
y = data[:, 2:3]

import tensorflow as tf

def sigmoid(el):
    return 1. / (1 + tf.exp(-el))

def plot_boundary():
    plt.subplot()
    plain_x = np.array(range(30, 100)).reshape(70, 1)
    boundary = (0.5 - np.matmul(np.concatenate((np.ones((70, 1)), plain_x), axis=1), np.array(sess.run(T))[:, 0:2].T)) / sess.run(T)[0][2]
    plt.plot(admitted[:, 0], admitted[:, 1], 'x')
    plt.plot(not_admitted[:, 0], not_admitted[:, 1], 'o')
    plt.plot(plain_x, boundary)
    plt.show()

costs = []

T = tf.Variable(tf.zeros([1, n + 1], dtype=tf.float64))
h = sigmoid(tf.matmul(x, tf.transpose(T)))

J = tf.reduce_sum(-y * tf.log(h) - (1 - y) * tf.log(1 - h)) / m

train = tf.assign(T, T - (alpha * tf.reduce_sum((h- y) * x, 0) / m))
init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)
for i in range(100000):
    sess.run(train)
    if i % 200000 == 0:
        plot_boundary()
    costs.append(sess.run(J))
print(sess.run(T))

plt.plot(costs)
```

```python
plot_boundary()
```

# With Feed-Dicts

```python
import tensorflow as tf

def sigmoid(el):
    return 1. / (1 + tf.exp(-el))

alpha = 0.0011
costs = []
X = tf.placeholder(tf.float64)
Y = tf.placeholder(tf.float64)
T = tf.Variable(tf.zeros([1, n + 1], dtype=tf.float64))
h = sigmoid(tf.matmul(X, tf.transpose(T)))

J = tf.reduce_sum(-Y * tf.log(h) - (1 - Y) * tf.log(1 - h)) / m

train = tf.assign(T, T - (alpha * tf.reduce_sum((h - Y) * X, 0) / m))

init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)

for i in range(100000):
    _, b = sess.run([train, J], { X: np.concatenate((np.ones((m, 1)), data[:, 0:2]), axis = 1), Y: data[:, 2:3] })
    costs.append(b)

plt.plot(costs)
```

# With normalization

```python
import tensorflow as tf

alpha = 0.91
means = data[:, 0:2].mean(axis=0)
stds = data[:, 0:2].std(axis=0)
raw_x = np.concatenate((np.ones((m, 1)), (data[:, 0:2] - means ) / stds), axis = 1)

def sigmoid(el):
    return 1. / (1 + tf.exp(-el))

costs = []
X = tf.placeholder(tf.float64)
Y = tf.placeholder(tf.float64)
T = tf.Variable(tf.zeros([1, n + 1], dtype=tf.float64))
h = sigmoid(tf.matmul(X, tf.transpose(T)))

J = tf.reduce_sum(-Y * tf.log(h) - (1 - Y) * tf.log(1 - h)) / m

train = tf.assign(T, T - (alpha * tf.reduce_sum((h - Y) * X, 0) / m))

init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)

for i in range(100):
    _, b = sess.run([train, J], { X: raw_x, Y: data[:, 2:3] })
    costs.append(b)

plt.plot(costs)
```

# With an optimizer

```python
import tensorflow as tf

alpha = 0.21
means = data[:, 0:2].mean(axis=0)
stds = data[:, 0:2].std(axis=0)
raw_x = np.concatenate((np.ones((m, 1)), (data[:, 0:2] - means ) / stds), axis = 1)

def sigmoid(el):
    return 1. / (1 + tf.exp(-el))

costs = []
X = tf.placeholder(tf.float64)
Y = tf.placeholder(tf.float64)
T = tf.Variable(tf.zeros([1, n + 1], dtype=tf.float64))
h = sigmoid(tf.matmul(X, tf.transpose(T)))

J = tf.reduce_sum(-Y * tf.log(h) - (1 - Y) * tf.log(1 - h)) / m

train = tf.train.MomentumOptimizer(alpha, 0.3).minimize(J)

init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)

for i in range(1000):
    _, b = sess.run([train, J], { X: raw_x, Y: data[:, 2:3] })
    costs.append(b)

plt.plot(costs)
```

# Predicting the score

```python
original_example = [45., 85.]
normalized_example = np.concatenate(([1], ((original_example - means) / stds))).reshape((1, 3))
prediction = sess.run(sigmoid(tf.matmul(normalized_example, tf.transpose(T))))
prediction
```

```python
misclassifications = sess.run(tf.reduce_sum(tf.abs(tf.round(h) - data[:, 2:3])), { X: raw_x, Y: data[:, 2:3] })
print "Accuracy: ", 1-misclassifications / m
```

# Regularized Logistic Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

m = 118
n = 2
alpha = 0.51

data = np.loadtxt(open("ex2data2.txt","rb"),delimiter=",")
accepted = data[data[:, 2] == 1, :]
rejected = data[data[:, 2] == 0, :]

plt.plot(accepted[:, 0], accepted[:, 1], 'x')
plt.plot(rejected[:, 0], rejected[:, 1], 'o')
```

```python
x1 = data[:, 0]
x2 = data[:, 1]

# Polynomial Feature Mapping
features = np.ones(shape=(m, 1))
for i in range(1, 7):
    for j in range(i + 1):
        new = (x1 ** (i - j)) * (x2 ** j)
        features = np.append(features, new.reshape((m, 1)), axis=1)

import tensorflow as tf

def sigmoid(el):
    return 1. / (1 + tf.exp(-el))

def train_regression(l):
    costs = []
    X = tf.placeholder(tf.float64)
    Y = tf.placeholder(tf.float64)
    T = tf.Variable(tf.zeros([1, 28], dtype=tf.float64))
    h = sigmoid(tf.matmul(X, tf.transpose(T)))

    J = tf.reduce_sum(-Y * tf.log(h) - (1 - Y) * tf.log(1 - h)) / m + l * tf.reduce_sum(tf.slice(T ** 2, [0, 1], [1, -1])) / 2 / m

    train = tf.train.MomentumOptimizer(alpha, 0.4).minimize(J)

    init = tf.initialize_all_variables()
    sess = tf.Session()
    sess.run(init)

    for i in range(10000):
        _, b = sess.run([train, J], { X: features, Y: data[:, 2:3] })
        costs.append(b)

    plt.plot(costs)
    return sess.run(T)
```

```python
def plot(T):
    plt.plot(accepted[:, 0], accepted[:, 1], 'x')
    plt.plot(rejected[:, 0], rejected[:, 1], 'o')
    syn_x1, syn_x2 = np.mgrid[-1:1.3:40j, -1:1.3:40j]
    syn_x = np.vstack([syn_x1.ravel(), syn_x2.ravel()])
    syn_x1 = syn_x.T[:, 0]
    syn_x2 = syn_x.T[:, 1]
    syn_features = np.ones(shape=(len(syn_x.T), 1))
    for i in range(1, 7):
        for j in range(i + 1):
            new = (syn_x1 ** (i - j)) * (syn_x2 ** j)
            syn_features = np.append(syn_features, new.reshape((len(syn_x.T), 1)), axis=1)
    init = tf.initialize_all_variables()
    sess = tf.Session()
    sess.run(init)
    res = np.append(syn_x.T, np.array(sess.run(sigmoid(tf.matmul(syn_features, tf.transpose(T))))).reshape((len(syn_x.T), 1)), axis=1)
    res_accepted = res[res[:, 2] >= 0.5, :]
    res_rejected = res[res[:, 2] < 0.5, :]
    plt.plot(res_accepted[:, 0], res_accepted[:, 1], 's', alpha=0.1)
    plt.plot(res_rejected[:, 0], res_rejected[:, 1], 's', alpha=0.1)
    plt.plot(accepted[:, 0], accepted[:, 1], 'x')
    plt.plot(rejected[:, 0], rejected[:, 1], 'o')
```

```python
t = train_regression(0.0)
```
