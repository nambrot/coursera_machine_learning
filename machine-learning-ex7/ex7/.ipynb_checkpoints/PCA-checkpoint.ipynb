```python
%matplotlib inline
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
import tensorflow as tf

mat = scipy.io.loadmat('ex7data1.mat')
# plt.plot(mat['X'][:, 0:1], mat['X'][:, 1:2], 'x')


npX = np.array(mat['X'])
normX = (npX - npX.mean(0)) / npX.std(0)
```

```python
m = len(npX)
X = tf.constant(normX, dtype=tf.float64)
sigma = tf.matmul(tf.transpose(X), X) / m
s, u, v = tf.svd(sigma)

init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)
sess.run(s)

u_reduce = tf.slice(u, [0, 0], [-1, 1])
z = tf.matmul(tf.transpose(u_reduce), tf.transpose(X))
X_rec = tf.matmul(tf.transpose(z), tf.transpose(u_reduce))
a = sess.run(z)
plt.plot(normX[:, 0], normX[:, 1], 'x')
# plt.plot(a, a, 'o')
b = sess.run(X_rec)
plt.plot(b[:, 0], b[:, 1], '^')
```

```python
faces = scipy.io.loadmat('ex7faces.mat')['X']
def plot_face(face):
    plt.imshow(face.reshape(32, 32), cmap='gray')
```

```python
npFaces = np.array(faces)
normFaces = (npFaces - npFaces.mean(0)) / npFaces.std(0)
m = 5000
X = tf.constant(normFaces, dtype=tf.float64)
sigma = tf.matmul(tf.transpose(X), X) / m
s, u, v = tf.svd(sigma)

init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)
sess.run(s)

# u_reduce = tf.slice(u, [0, 0], [-1, 1])
# z = tf.matmul(tf.transpose(u_reduce), tf.transpose(X))
# X_rec = tf.matmul(tf.transpose(z), tf.transpose(u_reduce))
```

```python
red = sess.run(tf.transpose(u))
plt.figure(1)
plt.subplot(251)
plot_face(red[10])
plt.figure(1)
plt.subplot(252)
plot_face(red[11])
plt.figure(1)
plt.subplot(253)
plot_face(red[12])
plt.figure(1)
plt.subplot(254)
plot_face(red[13])
plt.figure(1)
plt.subplot(255)
plot_face(red[14])
plt.figure(1)
plt.subplot(256)
plot_face(red[15])
plt.figure(1)
plt.subplot(257)
plot_face(red[16])
plt.figure(1)
plt.subplot(258)
plot_face(red[17])
plt.figure(1)
plt.subplot(259)
plot_face(red[18])
```

```python
# Reconstruction
u_reduce = tf.slice(u, [0, 0], [-1, 100])
z = tf.matmul(tf.transpose(u_reduce), tf.transpose(X))
X_rec = tf.matmul(tf.transpose(z), tf.transpose(u_reduce))
```
