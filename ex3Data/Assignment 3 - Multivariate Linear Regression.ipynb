```python
import numpy as np
```

```python
m = 47
n = 2
iterations = 50
alpha = 0.03
raw_x = np.loadtxt('ex3x.dat').astype(np.float64)
raw_y = np.loadtxt('ex3y.dat').astype(np.float64)

means = raw_x.mean(axis=0)
stds = raw_x.std(axis=0)
x = np.concatenate((np.ones((m, 1)), ((raw_x - means ) / stds)), axis = 1)
y = raw_y.reshape((m, 1))
```

```python
import tensorflow as tf
# Learning Rates to try
alphas = [0.01, 0.03, 0.1, 0.3, 1, 1.2]
# Plot this matrix
j0 = np.zeros((iterations, len(alphas)))

# Run
for alpha_index in range(len(alphas)):
    alpha = alphas[alpha_index]
    T = tf.Variable(tf.zeros([1, n + 1], dtype=tf.float64))
    Tt = tf.transpose(T)
    h0 = tf.matmul(x, Tt)
    cost =  tf.matmul(tf.transpose(tf.matmul(x, Tt) - y), (tf.matmul(x, Tt) - y))  / 2 / m
    train = tf.assign(T, T - tf.transpose(alpha * tf.matmul(tf.transpose(x), (h0 - y)) / m))
    init = tf.initialize_all_variables()
    # Launch the graph.
    sess = tf.Session()
    sess.run(init)

    for step in range(iterations):
        sess.run(train)
        j0[step][alpha_index] = sess.run(cost)
    %matplotlib inline
    import matplotlib.pyplot as plot
    plot.plot(j0)
    plot.legend(map(lambda x: "a: " + str(x), alphas))
```

```python
print("Theta: ", sess.run(T))
```

```python
print("Prediction for 1650 sqm and 3 BRs", sess.run(tf.matmul(np.concatenate((np.ones((1, 1)), (([[1650, 3]] - means ) / stds)), axis = 1), Tt)))
```

### Normal Equations

```python
import tensorflow as tf
sess = tf.Session()
xx = np.concatenate((np.ones((m, 1)), raw_x), axis = 1)
xt = tf.transpose(xx)
theta = tf.matmul(tf.matmul(tf.matrix_inverse(tf.matmul(xt, xx)), xt), y)
new_room = tf.constant([[1], [1650], [3]], dtype=tf.float64)

%matplotlib inline
import matplotlib.pyplot as plot
plot.plot(np.concatenate((sess.run(tf.matmul(xx, theta)), y), axis = 1))
plot.legend(['normal', 'y'])
```

```python
print("Prediction for 1650 sqm and 3 BRs", sess.run(tf.matmul(tf.transpose(theta), new_room)))
```
