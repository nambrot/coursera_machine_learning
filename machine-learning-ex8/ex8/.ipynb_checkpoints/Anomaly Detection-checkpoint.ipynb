```python
%matplotlib inline
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
import tensorflow as tf

mat = scipy.io.loadmat('ex8data1.mat')

npX = np.array(mat['X'])
means = npX.mean(0)
stds = npX.std(0)
```

```python
means
```

```python
stds
```

```python
x= np.arange(0,25)
y= np.arange(0,25)

def p(x, u, v):
    return np.exp(-(x - u)**2/v**2) / np.sqrt(2*np.pi) / v

X,Y = np.meshgrid(x,y)
Z = p(X, means[0], stds[0]) * p(Y, means[1], stds[1])
```

```python
CS = plt.contour(X, Y, Z, [0.0001, 0.001])
plt.clabel(CS, inline=1, fontsize=10)
plt.plot(npX[:, 0], npX[:, 1], 'x')
```
